```python
import math
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
file_path = files.upload()
```

```python
fileName = list(file_path)[0]
text = file_path[fileName].decode('utf-8')
texts = text.split("\r\n")
```

```python
x_values = []
for i in range(len(texts)) :
  x_values.append(1/22050 * i)
y_values = []
for i in range (len(texts)) :
  changed_float = float(texts[i])
  y_values.append(changed_float)
```

```python
plt.plot(x_values,y_values)
plt.show()
```
