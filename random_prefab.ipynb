```python
using System.Collections;
using System.Collections.Generic;
using UnityEngine;
using UnityEngine.SceneManagement;

public class iniateObj : MonoBehaviour
{
    public GameObject[] prefabs;
    public GameObject final;
    public GameObject stop;
    public GameObject sky;
    public GameObject ground;
    public GameObject mainchar;
    private float spawnRangeX = 8.20f;
    public float SceneTime = 0;
    public float EndTime = 30;
    public bool GameEnd = false;
    public bool initrue = false;

    private void Start()
    {

        sky = GameObject.Find("sky");
        ground = GameObject.Find("ground");
        mainchar = GameObject.Find("maincharacter01");
        InvokeRepeating("RepeatSpawn", 2, 0.7f);
    }
       

    void RepeatSpawn()
    {
        Vector3 spawnPos = new Vector3(Random.Range(-spawnRangeX, spawnRangeX), 7, 0);
        int prefabINDEX = Random.Range(0, prefabs.Length);
        Instantiate(prefabs[prefabINDEX], spawnPos, Quaternion.identity);
    }

    void Update() {
        Vector3 spawn = new Vector3(0, 0, 0);
        if (GameEnd == true)
        {
            CancelInvoke();
            Destroy(sky);
            Destroy(ground);
            Destroy(mainchar);
            if (initrue == false)
            {
                Instantiate(final, spawn, Quaternion.identity);
                initrue = true;
            }
        }
        else
        {
            SceneTime += Time.deltaTime;
            if (SceneTime >= EndTime)
            {           
                Instantiate(stop, spawn, Quaternion.identity);
                if (SceneTime >= EndTime + 2)
                {
                    GameEnd = true;
                }
            }
        }    
    }
}
```
