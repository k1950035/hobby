```python
%matplotlib inline
from google.colab import files
import pandas as pd
import io
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from konlpy.tag import Okt 
from numpy import dot 
from numpy.linalg import norm
import operator 
```

```python
#크롤링을 통해서 기사를 수집
#자연어 처리 이용해서 명사 빈도수 파악(주제 찾기)
#문장 유사도 이용 -> 문장의 유사도 파악 뒤, 해당 문장 출력.
```

```python
def load_link(lk,a) :
  global link
  htmls = requests.get(lk+str(a))
  news = htmls.text
  soup = BeautifulSoup(news, 'html.parser')
  find_href = soup.find('ul',{"class":"list"}).find_all("li",{"class" : False})

  for href in find_href :
    link.append(href.find("a")["href"])

def analyze_sentence(hrefs) :
  https_link = "https:"+hrefs
  htmls = requests.get(https_link)
  news = htmls.text
  soup = BeautifulSoup(news, 'html.parser')

  sentence = []

  sentence.append(soup.find('h1',{"class":"tit"}).get_text())
  for p in soup.find_all('p',{"class":False}) :
    sentence.append(p.get_text())
  sentence = ''.join(sentence)
  return sentence

def analyze_noun(sentence) :
  okt = Okt()
  noun = okt.nouns(sentence)

  for i,v in enumerate(noun) :
    if len(v)<2:
      noun.pop(i)
  for i,v in enumerate(noun) :
    if len(v)<2:
      noun.pop(i)
  count = Counter(noun)
  noun_list = count.most_common(100)
  main_keyword = noun_list[0][0]


  return main_keyword

def count_keyword(main_keyword) :
  count = {}
  for i in main_keyword :
    try: count[i] += 1
    except: count[i] = 1

  count = sorted(count.items(),reverse=True,key=operator.itemgetter(1))
  return count
```

```python
link_page = 5
lk = "https://www.yna.co.kr/industry/technology-science/"
link = []
main_keyword = []

for i in range(link_page) :
  load_link(lk,i)

for i in range(len(link)) :
  sentence = analyze_sentence(link[i])
  main_keyword.append(analyze_noun(sentence))
```

```python
count = count_keyword(main_keyword)
print(count)
```
