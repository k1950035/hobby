```python
import numpy as np
from google.colab import files
from collections import Counter
from konlpy.utils import pprint
from konlpy.tag import Kkma
from konlpy.tag import Okt 
from numpy import dot 
from numpy.linalg import norm 
import numpy as np
```

```python
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))

''' 파일을 직접 선택하면 해당 파일의 모든 라인을 리스트 형태로 반환'''
def load_file_lines():
  myfile = files.upload()
  fileName = list(myfile)[0]
  text = myfile[fileName].decode('utf-8')
  texts = text.split("\r\n")
  print(texts)
  return texts 

def make_matrix(feats, list_data) :
  freq_list= []
  for feat in feats :
    freq = 0
    for word in list_data :
      if feat == word :
        freq +=1
    freq_list.append(freq)
  return freq_list

def analyze_morphs(sentences):
    vector_text = []
    feats=[]

    for sentence in sentences :
      morphs = okt.morphs(sentence)
      vector_text.append(morphs)
      feats+=morphs
    feats = list(set(feats))
    return feats, vector_text

def one_hot(total_feats, sentece):
  return np.array(make_matrix(total_feats, sentece))

def one_hot_total(total_feats, sentences):
  vector_array = []
  for sentence in sentences :
    vector_array.append(one_hot(total_feats, sentence))
  return vector_array

def get_similarity(sentences, user_sentence):
  # 파일 정보 처리
  feats, morphs_text = analyze_morphs(sentences)
  onehots = one_hot_total(feats, morphs_text)

  # 유저 입력 처리

  input_feats, input_vector_text = analyze_morphs(user_sentence)
  user_onehot = one_hot_total(feats, input_vector_text)[0]

  # 최대 유사도 결과
  return max(list(map(lambda x : cos_sim(x, user_onehot), onehots)))
```

```python
# 형태소 분석 객체 생성
okt = Okt()

# 파일 로드
sentences = load_file_lines()
```

```python
user_sentence = input("대사를 입력 :")
get_similarity(sentences, user_sentence)
```
