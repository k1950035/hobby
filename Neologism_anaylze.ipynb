```python
def load_file() : #파일을 불러온다.
  myfile = files.upload()
  fileName = list(myfile)[0]
  text = myfile[fileName].decode('utf-8')
  texts = text.split("\r\n") #개행문자 제거
  return texts
 
data_array = load_file()
print(data_array)
```

```python
def make_data_array(data_array) : #데이터 배열(신조어,뜻풀이)을 만든다.
  for i in range(len(data_array)) :
    texts_value = data_array[i].split(";;;") #;;;을 기준으로 문장을 쪼갠다. [신조어,신조어뜻]이 texts_value에 담긴다.
    text_array.append(texts_value[0]) #text_value의 0번째(신조어)를 text_array 배열에 담는다.
    text_mean.append(texts_value[1]) #text_value의 1번째(신조어 뜻)를 text_mean 배열에 담는다.
  return text_array,text_mean 

text_array = [] #신조어를 담는 배열
text_mean = [] #신조어의 뜻을 담는 배열

text_array,text_mean = make_data_array(data_array) #text_array, text_mean에 신조어,신조어뜻 배열을 반환받는다.

print(text_array)
print(text_mean)
```

```python
def analyze_Syllable(sentences): #한 단어씩 분리
  end_spot = ['.','?','!',',',';','~']
  vector_text = []
  feats=[]
  for sentence in sentences :
    if sentence in end_spot : #마침표 제거
      pass
    else :
      vector_text.append(sentence) #나누어진 음절 vector_text에 추가한다.
  for i in range(len(vector_text)) :
    feats.append(vector_text[i][0])
  return feats, vector_text 
```

```python
def make_matrix(feats, list_data) : #원핫 인코딩 배열 생성
  freq_list= []
  for feat in feats :
    freq = 0
    for word in list_data :
      if feat == word :
        freq +=1 
    freq_list.append(freq) #feat와 word가 일치할 경우, 1을 더한다.
  return freq_list

def one_hot(total_feats, sentece): #원핫 인코딩 수행
  return np.array(make_matrix(total_feats, sentece))

def one_hot_total(total_feats, sentences):
  vector_array = [] 
  for sentence in sentences :
    vector_array.append(one_hot(total_feats, sentence)) 
  return vector_array

def cos_sim(A, B): #코사인 유사도 공식(1에 가까울 수록 두 벡터 사이의 유사도가 높다.)
  return dot(A, B)/(norm(A)*norm(B))

def get_similiarity(split_input_sentence, text_array) :#유사도 구하기
  text_final_sim = []
  for j in range(len(split_input_sentence)) :
    #txt_ar_vector_text = 신조어 데이터셋  vector_text = 입력문장
    text_sim=[]
    feats, vector_text = analyze_Syllable(split_input_sentence[j]) #입력 문장의 형태소 분석
    onehots = one_hot_total(feats, vector_text) #입력 문장 : 원-핫 인코딩 과정을 거친다.
    for i in range(len(text_array)) :
      max_sim = 0 
      txt_ar_feats, txt_ar_vector_text = analyze_Syllable(text_array[i]) #신조어 데이터셋의 형태소 분석
      while(len(txt_ar_vector_text)<len(vector_text)) : #두 배열의 길이를 맞춰줌
        txt_ar_vector_text.append(" ")
      if(len(txt_ar_vector_text)>len(vector_text)) : #신조어의 길이가 입력 단어의 길이보다 긴 경우, 유사도를 0으로 처리
        text_sim.append(0)
      else :
        user_onehot = one_hot_total(feats, txt_ar_vector_text)
        for e in range(len(txt_ar_vector_text)) :
          sim = max(cos_sim(onehots,user_onehot)[e])
          max_sim+=sim
        text_sim.append(max_sim) #유사도가 있다면, 그 유사도를 text_sim에 append함
    text_final_sim.append(text_sim)
  return text_final_sim #최종적으로 모든 유사도를 담은 배열을 반환함  
```

```python
def print_sentence(sim_array,split_input_sentence) :
  pos = [] #유사도, 뜻의 인덱스, 해석이 위치할 자리
  for i in range(len(sim_array)) :
    max_value = -1
    max_index = -1
    for value in sim_array[i] :
      if value == value : #값이 존재하는 경우(nan처리)
        if(value>max_value and value>=0.65) : #유사도는 0.65 이상인 경우를 카운트
          max_value = value
          max_index = sim_array[i].index(max_value)
          pos.append([max_value,max_index,i])
  
  for i in range(len(pos)) :
    word_pos = pos[i][2] #해석이 위치할 자리
    word_sim = pos[i][0] #해당 신조어와의 유사도
    word_mean = pos[i][1] #해당 신조어의 text_mean에서의 인덱스
    split_input_sentence.insert(word_pos+(i+1),"("+text_mean[word_mean]+")") #해석이 위치할 자리에 신조어의 뜻 추가

  print("pos : ",pos) #포스 출력
  print("변형된 문장 : ",listToString(split_input_sentence)) #최종 결과 출력

def listToString(str_list) : #리스트를 문장으로 변환 후 출력
  result = ""
  for s in str_list :
    result += s+" "
  return result.strip()
```

```python
input_sentence = input("문장을 입력하세요 :")
split_input_sentence = input_sentence.split(" ") #띄어쓰기를 기준으로 분할    

sim_array = get_similiarity(split_input_sentence, text_array)
#유사도 구하기
print_sentence(sim_array,split_input_sentence)
#설명이 추가된 문장 출력
```
